So let us find out some missing data below

```python
# Highlighting some of missing row
df_missing = df.head(5)
def highlight_last_row(s):
    return ['background-color: #ff008c' if i==len(s)-3  else '' for i in range(len(s))]

ms = df_missing.style.apply(highlight_last_row)
ms
```

```python
def missing_data(df):
    '''
    Find the missing data in the dataset and return the percentage
    of the missing data
    '''
    missing_data = df.isnull().sum().to_frame()
    missing_data['Percentage'] =  (df.isnull().sum()/df.shape[0])*100
    missing_data.columns = ['Count_of_Missing_data','Percentage']
    missing_data = missing_data.round(1)
    return missing_data
```

```python
missing_data(df)
```
